```python
#Data Cleaning
import pandas as pd
bigdata = pd.read_csv('employee-compensation.csv')
bigdata.head()
```

```python
#Memilih data gaji pegawai yang bukan bernilai 0
datapegawai = bigdata[bigdata.Salaries != 0]
#Membuat funsgi ABS 
salaries={}
for baris in datapegawai['Salaries']:
    if baris < 0 :
        salaries=abs(datapegawai['Salaries'])
overtime={}
for baris in datapegawai['Overtime']:
    if baris < 0 :
        overtime=abs(datapegawai['Overtime'])
other_salaries={}
for baris in datapegawai['Other Salaries']:
    if baris < 0 :
        other_salaries=abs(datapegawai['Other Salaries'])
total_salaries={}
for baris in datapegawai['Total Salary']:
    if baris < 0 :
        total_salaries=abs(datapegawai['Total Salary'])
retirement={}
for baris in datapegawai['Retirement']:
    if baris < 0 :
        retirement=abs(datapegawai['Retirement'])
health_and_dental={}
for baris in datapegawai['Health and Dental']:
    if baris < 0 :
        health_and_dental=abs(datapegawai['Health and Dental'])
other_benefit={}
for baris in datapegawai['Other Benefits']:
    if baris < 0 :
        other_benefit=abs(datapegawai['Other Benefits'])
total_benefit={}
for baris in datapegawai['Total Benefits']:
    if baris < 0 :
        total_benefit=abs(datapegawai['Total Benefits'])
total_compensation={}
for baris in datapegawai['Total Compensation']:
    if baris < 0 :
        total_compentation=abs(datapegawai['Total Compensation'])
```

```python
#Membuat data Frame Baru
datasuci=pd.DataFrame({'salaries': salaries, 'overtime': overtime,'retirement':retirement,'other salaries': other_salaries,'total salary': total_salaries,'health and dental': health_and_dental,'other benefits': other_benefit,'total compensation': total_compensation})
```

```python
datanew=datapegawai.drop('Salaries',axis=1)
datanew=datanew.drop('Overtime',axis=1)
datanew=datanew.drop('Other Salaries',axis=1)
datanew=datanew.drop('Total Salary',axis=1)
datanew=datanew.drop('Retirement',axis=1)
datanew=datanew.drop('Health and Dental',axis=1)
datanew=datanew.drop('Other Benefits',axis=1)
datanew=datanew.drop('Total Benefits',axis=1)
datanew=datanew.drop('Total Compensation',axis=1)
datasuciterbaru=pd.concat([datanew, datasuci],axis=1)
datasuciterbaru.head()
```

```python
datasuci=datasuciterbaru.drop_duplicates(keep=False)
```
