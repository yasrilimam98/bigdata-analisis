```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
datasuci = pd.read_csv('datasuci.csv')
datasuci.head()
```

```python
#membuat ukuran gambar
plt.figure(figsize=(12,8))
plt.title('Kompensasi gaji pegawai di Kota fransisco (USD)', fontdict={'fontweight':'bold','fontsize':18})
plt.plot(datasuci.index, datasuci.salaries, 'b.-', label='Gaji pokok')
plt.plot(datasuci.index, datasuci.overtime, 'r.-', label='Gaji lembur')
plt.plot(datasuci.index, datasuci.retirement, 'g.-', label='Gaji pensiun')
plt.plot(datasuci.index, datasuci['other salaries'], 'm.-', label='Gaji tambahan')
plt.plot(datasuci.index, datasuci['other benefits'], 'c.-', label='Tunjang lain')
plt.plot(datasuci.index, datasuci['health and dental'], 'k.-', label='Tunjangan Kesehatan')

plt.xlabel('Tahun')
plt.ylabel('Dolar USD')

plt.legend()
plt.show()
```
